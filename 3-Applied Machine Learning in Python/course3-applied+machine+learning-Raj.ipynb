```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

#print(cancer.DESCR) # Print the data set description
```

```python
def answer_one():
    
    df=pd.DataFrame(cancer['data'],cancer['target'])
    df.columns=cancer['feature_names']
    df=df.assign(target=cancer['target'])
    
    return df


answer_one()
```

```python
def answer_two():
    cancerdf = answer_one()
    malignant=0
    benign=0
    for x in cancerdf['target']:
        if x==0:
            malignant +=1
        else:
            benign +=1
    
    target=pd.Series([malignant,benign])
    
    return target


answer_two()
```

```python
def answer_three():
    cancerdf = answer_one()
    X=pd.DataFrame(cancerdf.loc[:,'mean radius':'worst fractal dimension'])
    y=pd.Series(cancerdf['target'])
    # Your code here
    
    return X, y
answer_three()
```

```python
from sklearn.model_selection import train_test_split

def answer_four():
    X, y = answer_three()
    
    X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=143)
    
    return X_train, X_test, y_train, y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier

def answer_five():
    X_train, X_test, y_train, y_test = answer_four()
    knn = KNeighborsClassifier(n_neighbors = 1)
    
    return knn.fit(X_train,y_train)
```

```python
def answer_six():
    cancerdf = answer_one()
    means = cancerdf.mean()[:-1].values.reshape(1, -1)
    knn=answer_five()
    prediction = np.array(knn.predict(means))
    
    return prediction
```
